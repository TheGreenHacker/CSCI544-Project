```python
final = pd.read_excel(path_to_data)
```

```python
path_to_text_data = list(final['Review File'])
```

```python
text_list = []
for i in range(0, len(path_to_text_data)):
  with open('/content/gdrive/MyDrive/Colab Notebooks/topics/' + path_to_text_data[i], "r+", encoding = 'utf8', errors = 'ignore') as file:
    text = file.read()
    text = text.replace('\n', '')
    text_list.append(text)
```
